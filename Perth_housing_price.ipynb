```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 업로드
from google.colab import files
uploaded = files.upload()

# CSV 불러오기
df = pd.read_csv("all_perth_310121 2.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

# suburb별 평균 가격
mean_price = df.groupby("SUBURB")["PRICE"].mean().sort_values(ascending=False)

# 상위 10개 suburb만
mean_price.head(10).plot(kind='bar', figsize=(10,6), color='skyblue')
plt.ylabel("Average Price (AUD)")
plt.title("Top 10 Suburbs by Average Price")
plt.show()
```

```python
mean_price_by_bed = df.groupby("BEDROOMS")["PRICE"].mean()
mean_price_by_bed.plot(kind='bar', color='salmon')
plt.title("Average Price by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price (AUD)")
plt.show()
```

```python
plt.scatter(df["GARAGE"], df["PRICE"], alpha=0.5)
plt.title("Garage vs Price")
plt.xlabel("Garage Spaces")
plt.ylabel("Price (AUD)")
plt.show()
```

```python
plt.scatter(df["FLOOR_AREA"], df["PRICE"], alpha=0.5, color='green')
plt.title("Floor Area vs Price")
plt.xlabel("Floor Area (sqm)")
plt.ylabel("Price (AUD)")
plt.show()
```

```python
print(f"Average price: {df['PRICE'].mean():,.0f} AUD")
print(f"Most common number of bedrooms sold: {df['BEDROOMS'].mode()[0]}")
```
